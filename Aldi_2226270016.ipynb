```python
# Import library
import urllib.request
import cv2 # https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html
import matplotlib.pyplot as plt
import numpy as np
```

# a. Memuat Gambar Grayscale dari Link

```python
# Get image from link google image
urllib.request.urlretrieve(
'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSfh5OB__yeqZR9DVocYvj6-aAym35c_yHWfQ&s.jpg', # link from google image
"Gambar_1_Sunset.png")
```

```python
# Load image
image_1 = cv2.imread('Gambar_1_Sunset.png') # cv2.IMREAD_GRAYSCALE

# Dapatkan resolusi gambar dalam piksel
width, height,channel = image_1.shape
print(f"Resolusi Gambar 1: {width}x{height}x{channel} piksel")
```

```python
# Plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 1, 1)
plt.imshow(image_1)
plt.title("Gambar 1 Sunset")
plt.axis('off')

plt.show()
```

```python
# Dapatkan dimensi gambar yang lebih kecil
min_height = min(image_1.shape[0], image_1.shape[1])

# Plot figure
plt.figure(figsize=(10, 5))
plt.subplot(1, 1, 1) # Changed the subplot index to 1
plt.imshow(image_1, cmap='gray')
plt.title('Gambar 1 Sunset')

plt.show()
```

# b. Histogram Equalization

```python
gray_image = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)

# Lakukan histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Menampilkan gambar yang original dan gambar equalized
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Gambar 1 Sunset Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Gambar 1 Sunset Equalized')
plt.axis('off')

plt.show()

# Melakukan menghitung dan memplot gambar histogram
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(gray_image.ravel(), 256, [0, 256])
plt.title('Gambar 1 Sunset Original')

plt.subplot(1, 2, 2)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Gambar 1 Sunset Histogram Equalized ')

plt.show()
```

# c. Filtering di Domain Frekuensi menggunakan Gaussian Filtering

```python
# Mengubah Gambar ke Domain Frekuensi
F = np.fft.fft2(gray_image)
Fshift = np.fft.fftshift(F)

# Plot figure
plt.figure(figsize=(5,5))
plt.imshow(np.log1p(np.abs(F)), cmap='gray')
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(np.log1p(np.abs(Fshift)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Import library
import urllib.request
import cv2 # https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html
import matplotlib.pyplot as plt
import numpy as np
```

# a. Memuat Gambar Grayscale dari Link

```python
# Get image from link google image
urllib.request.urlretrieve(
'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQ9_ZWypsp_cJSTdtukGtvYAM0rSen1Ta5iSw&s.jpg', # link from google image
"Gambar_2_Kucing.png")
```

```python
# Load image
image_2 = cv2.imread('Gambar_2_Kucing.png') # cv2.IMREAD_GRAYSCALE

# Dapatkan resolusi gambar dalam piksel
width, height,channel = image_2.shape
print(f"Resolusi Gambar 2: {width}x{height}x{channel} piksel")

# Plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 1, 1)
plt.imshow(image_2)
plt.title("Gambar 2 Kucing")
plt.axis('off')

plt.show()
```

```python
# Dapatkan dimensi gambar yang lebih kecil
min_height = min(image_2.shape[0], image_2.shape[1])

# Plot figure
plt.figure(figsize=(10, 5))
plt.subplot(1, 1, 1) # Changed the subplot index to 1
plt.imshow(image_2, cmap='gray')
plt.title('Gambar 2 Kucing')

plt.show()
```

```python
ref = cv2.imread('Gambar_2_Kucing.png')

plt.imshow(ref), plt.grid(False)

# Melakukan pemfilteran spasial
plt.title('Gambar 2 Kucing Original')
plt.xticks([])
plt.yticks([])
plt.show()
```

# b. Spatial Filtering

```python
# Mendefinisikan kernel menggunakan numpy.
kernel_7 = np.ones((7,7),np.float32)/49
kernel_5 = np.ones((5,5),np.float32)/25
kernel_3 = np.ones((3,3),np.float32)/9

# Menggabungkan gambar dengan kernel.
# -1 yang berarti pusat kernel terletak pada piksel tengah.
# Membandingkan dua ukuran kernel.
filtered_7 = cv2.filter2D(ref,-1,kernel_7)
filtered_5 = cv2.filter2D(ref,-1,kernel_5)
filtered_3 = cv2.filter2D(ref,-1,kernel_3)

#  Memplotkan yang hasilnya dalam dua subplot
fig=plt.figure(figsize=(14, 14), dpi= 80, facecolor='w', edgecolor='k')

plt.subplot(131), plt.imshow(filtered_3), plt.title('3x3 Filter Kucing')
plt.grid(False)
plt.xticks([])
plt.yticks([])

plt.subplot(132), plt.imshow(filtered_5), plt.title('5x5 Filter Kucing')
plt.grid(False)
plt.xticks([])
plt.yticks([])

plt.subplot(133), plt.imshow(filtered_7), plt.title('7x7 Filter Kucing')
plt.grid(False)
plt.xticks([])
plt.yticks([])


plt.show()
```

#c. Tampilkan dan Bandingkan Hasil
**Penjelasan** :                                        

Averaging filter adalah berfungsi sebagai untuk menghaluskan gambar yang menjadi kurang detail atau blur pada gambar tersebut. Intensitas rata-rata piksel di sekitarnya dihitung oleh filter ini, lalu nilai piksel tengah diubah menjadi nilai rata-rata tersebut.
Pada soal ini melakukan pemfilteran yaitu 3x3, 5x5 dan 7x7.

- Pemfilteran 3x3 : yang terjadi adalah pada gambar nya masih terlihat jelas sama seperti dengan gambar awal atau gambar aslinya tetapi sudah terjadi perubahan pada gambar tersebut baik dari segi warna.

- Pemfilteran 5x5 : yang terjadi pada filter 5x5 adalah gambar sedikit lebih ngeblur dibandingkan dengan gambar sebelumnya atau gambar pada filter 3x3.

- Pemfilteran 7x7 : yang terjadi pada pemfilteran 7x7 adalah gambar sudah terlihat ngeblur dibandingkan gambar sebelum-seblumnya tetapi masih terlihat, perbedaan sangat terasa pada filter 7x7 ini dikarenakan sangat ngeblur dibandingkan dengan gambar aslinya.

**Kesimpulan**

Jadi, kesimpulan nya adalah averaging filter berfungsi sebagai untuk menghaluskan gambar atau mengeblur pada gambar, semakin banyak dan semakin besar nilai kernel averaging filter, maka akan menghasilkan gambar yang semakin blur dengan sangat parah sampai sulit untuk dilihat dalam kondisi jauh.
