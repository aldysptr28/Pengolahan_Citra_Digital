```python
# libraries
import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# orginal image
urllib.request.urlretrieve('https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png', "Lenna.png")
```

```python
img = cv2.imread('Lenna.png', cv2.IMREAD_GRAYSCALE)
img = img/255
```

```python
plt.imshow(img, cmap='gray')
plt.title("Gambar Lenna")
plt.show()
```

```python
# blank image
x,y = img.shape
g = np.zeros((x,y), dtype=np.float32)
# salt and pepper amount
pepper = 0.1
salt = 0.95
```

```python
for i in range(x):
    for j in range(y):
        rdn = np.random.random()
        if rdn < pepper:
            g[i][j] = 0
        elif rdn > salt:
            g[i][j] = 1
        else:
            g[i][j] = img[i][j]

# If using in a standard Python environment with OpenCV installed:
# cv2.imshow('Image with Salt and Pepper Noise', g)

# If using in Google Colab or a similar environment:
from google.colab.patches import cv2_imshow # Import cv2_imshow from google.colab.patches
cv2_imshow(g) # Use cv2_imshow to display the image


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# (optional) save the image
from skimage import img_as_ubyte
cv2.imwrite('5percent.jpg', img_as_ubyte(g))
```
